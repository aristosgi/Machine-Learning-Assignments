```python
#get the values i will use
ols_1_values = pd.DataFrame(ols_1.tables[1])[[0,1,2]]
ols_2_values = pd.DataFrame(ols_2.tables[1])[[0,1,2]]
ols_3_values = pd.DataFrame(ols_3.tables[1])[[0,1,2]]


#Add variable fot the dataframe
ols_1_values.loc[len(ols_1_values.index)] = ['Controls', '✖' , '']
ols_2_values.loc[len(ols_2_values.index)] = ['Controls', '✓' , '']
ols_3_values.loc[len(ols_3_values.index)] = ['Controls', '✓' , '']

ols_1_values.loc[len(ols_1_values.index)] = ['Zone FE', '✖' , '']
ols_2_values.loc[len(ols_2_values.index)] = ['Zone FE', '✖' , '']
ols_3_values.loc[len(ols_3_values.index)] = ['Zone FE', '✓' , '']

#AIC
ols_1_values.loc[len(ols_1_values.index)] = ['AIC', first_.aic , '']
ols_2_values.loc[len(ols_2_values.index)] = ['AIC',second_.aic  , '']
ols_3_values.loc[len(ols_3_values.index)] = ['AIC',  third_.aic, '']

#WALD X^2
ols_1_values.loc[len(ols_1_values.index)] = ['Wald x^2',first_.wald_test(['hostcitytime=0' ,'hostcitytime2=0','hostcity=0' ,'time=0','time2=0']).statistic[0][0]  , '']
ols_2_values.loc[len(ols_2_values.index)] = ['Wald x^2',second_.wald_test(['hostcitytime=0' ,'hostcitytime2=0','hostcity=0' ,'time=0','time2=0','lnpop_1970=0','vote_frejuli=0','literacy_avg=0','lnrebact1974=0','lnrepression70_77=0']).statistic[0][0]  , '']
ols_3_values.loc[len(ols_3_values.index)] = ['Wald x^2',third_.wald_test(['hostcitytime=0' ,'hostcitytime2=0','hostcity=0' ,'time=0','time2=0','lnpop_1970=0','vote_frejuli=0','literacy_avg=0','lnrebact1974=0','lnrepression70_77=0' , 'zone2=0' , 'zone3=0' , 'zone4=0' , 'zone5=0']).statistic[0][0]  , '']

#r^2 , F statistic , Number of observations
r_sq_1 = ols_1.tables[0][0][3]
r_sq_2 = ols_2.tables[0][0][3]
r_sq_3 = ols_3.tables[0][0][3]
f_1 = ols_1.tables[0][2][3]
f_2 = ols_2.tables[0][2][3]
f_3 = ols_3.tables[0][2][3]
obs1 = ols_1.tables[0][5][1]
obs2 = ols_2.tables[0][5][1]
obs3 = ols_3.tables[0][5][1]

ols_1_values.loc[len(ols_1_values.index)] = ['F statistic', f_1 , '']
ols_2_values.loc[len(ols_2_values.index)] = ['F statistic', f_2 , '']
ols_3_values.loc[len(ols_3_values.index)] = ['F statistic', f_3 , '']
ols_1_values.loc[len(ols_1_values.index)] = ['R^2', r_sq_1 , '']
ols_2_values.loc[len(ols_2_values.index)] = ['R^2', r_sq_2 , '']
ols_3_values.loc[len(ols_3_values.index)] = ['R^2', r_sq_3 , '']
ols_1_values.loc[len(ols_1_values.index)] = ['Observations', obs1 , '']
ols_2_values.loc[len(ols_2_values.index)] = ['Observations', obs2 , '']
ols_3_values.loc[len(ols_3_values.index)] = ['Observations', obs3 , '']

#set index

ols_1_values = ols_1_values.set_index(0)
ols_2_values = ols_2_values.set_index(0)
ols_3_values = ols_3_values.set_index(0)

#Set column names

ols_1_values.columns = ols_1_values.iloc[0]
ols_1_values = ols_1_values.drop(ols_1_values.index[0])

ols_2_values.columns = ols_2_values.iloc[0]
ols_2_values = ols_2_values.drop(ols_2_values.index[0])

ols_3_values.columns = ols_3_values.iloc[0]
ols_3_values = ols_3_values.drop(ols_3_values.index[0])
```

Linear regression and predict values

```python
third_ = smf.ols("repression ~ hostcitytime + hostcitytime2 + hostcity + time + time2 + lnpop_1970 + vote_frejuli + literacy_avg + lnrebact1974 + lnrepression70_77 + zone2 + zone3 + zone4 + zone5", data=main_data)
results = third_.fit(cov_type='hc1')
#values from host cities before the tournament ends
a_ = main_data.loc[main_data['hostcity']==1]
a_ = a_.loc[a_['date'] <= ev]
a = results.predict(a_)

#values from non host cities before the tournament ends
b_ = main_data.loc[main_data['hostcity']!=1]
b_ = b_.loc[b_['date'] <= ev]
b = results.predict(b_)

```

##### Interpretation

In the Figure 5 we can see 2 plots.

* In the first plot the black line represents the predicted number of daily repression events in host cities for the three months leading up to and during the Word Cup. As we can see, this numbers increase for the first two months and then starting to decline.That indicates that the repression is higher a increasing prior to the world cup and decreases as the tournament approaches.

* In the second plot the black line represents the predicted number of daily repression events in non host cities.As we can see,the predicted number of daily repression events is extremly low compare with this in host cities.

The red dashed line is the average number of repression events include in both host an non host cities.As we can see the repression events in the host cities is above the average and in the non host cities the repression is under the average.

The blue line in both diagrams is the average of its situation(in the first the average in host cities and in the second in non host cities.

In the first diagramm the repression events that take place approximately during the 2 months before the Word Cup are above the average and the others are under the average.On the other hand in the second diagramm the situation is completly different.

A possible explanation is that the regime was trying to prevent any embarassment during the turnament
by controlling the population with violence before the tournament starts.
